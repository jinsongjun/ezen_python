```python
url = 'http://apis.data.go.kr/1160100/service/GetFnCoBasiInfoService/getFnCoOutl'

params = {
    'serviceKey' : 'dtbWOdJ/Cz5HE0DGLU+CRPe7pOW0NIQBUcGEqsHZaTRiYCI/5+zugwzQjcvuId7NPdg6rUiW+ft3fm7yqyD4pw==', 
    'pageNo' : '1', 
    'numOfRows' : '10', 
    'resultType' : 'xml'
}

response = requests.get(url, params=params)
```

```python
# xml 데이터를 dict 형태로 변환하기 위한 라이브러리 설치 
!pip install xmltodict
```

```python
import xmltodict
```

```python
data = xmltodict.parse(response.content)

data
```

```python
data2 = data['response']['body']['items']['item']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data2)
```

```python
df.columns
```

```python
# 데이터프레임을 csv로 저장(데이터의 형태를 데이터프레임에서 csv의 형태로 변환)
df.to_csv('금융회사기본정보.csv', index=False)
```
