```python
import pandas as pd
import yfinance as yf

# 주식 티커 리스트 설정
tickers = ["274400.KQ", "378340.KQ", "420770.KQ", "454910.KS", "463050.KS", "274400.KQ", "AMD", "GOOGL", "INTC", "MSFT", "NVDA", "SCHD", "TLT", "TSLA", "445090.KQ" ]

# 데이터 저장할 엑셀 파일 객체 생성
close_data = pd.DataFrame()

def get_close_prices(tickers):
    global close_data  # close_data를 전역 변수로 사용
    for ticker in tickers:
        # 데이터 다운로드
        stock_data = yf.download(ticker, start="2023-11-01", end="2024-03-01")
        
        # Close 컬럼 추출
        close = stock_data["Close"]
        
        # 컬럼 이름 설정
        close.columns = [ticker]
        
        # 데이터프레임에 추가
        if close_data.empty:
            close_data = close.to_frame(name=ticker)
        else:
            close_data = pd.concat([close_data, close.to_frame(name=ticker)], axis=1)
            
close_data = close_data.sort_index()            

get_close_prices(tickers)



```

```python
import pandas as pd
import pandas_datareader as pdr

# 회사채 티커 설정
tickers = ['KR6003491050']  # AAA 등급 회사채, BAA 등급 회사채

# 데이터 다운로드 기간 설정
start_date = '2023-11-01'
end_date = '2024-03-01'

# 빈 데이터프레임 생성
corp_bond_data = pd.DataFrame()

def get_corp_bond_prices(tickers, start_date, end_date):
    global corp_bond_data  # corp_bond_data를 전역 변수로 사용
    for ticker in tickers:
        # 데이터 다운로드
        bond = pdr.DataReader(f'BAMLC0A{ticker}YIEYLD', 'fred', start_date, end_date)
        
        # 데이터프레임에 추가
        if corp_bond_data.empty:
            corp_bond_data = bond['VALUE'].to_frame(name=ticker)
        else:
            corp_bond_data = pd.concat([corp_bond_data, bond['VALUE'].to_frame(name=ticker)], axis=1)

get_corp_bond_prices(tickers, start_date, end_date)

# 날짜 순으로 인덱스 정렬
corp_bond_data = corp_bond_data.sort_index()

# 결과 출력
print(corp_bond_data)
```
